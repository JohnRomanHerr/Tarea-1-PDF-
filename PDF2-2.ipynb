```python
import scipy.stats as ss
import numpy as np
import matplotlib.pyplot as plt
mu = 70
sigma = 2
```

```python
X = ss.norm(mu,sigma)
x=np.random.normal(mu, sigma, 10000)
```

```python
plt.plot(x,X.pdf(x),"r")
plt.show()
```

```python
datos = x
cuenta, cajas, ignorar = plt.hist(datos,50)
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma')
plt.text(70, .21, r'$\mu=70,\ \sigma=2$')
plt.show()
```

```python
x_1 = np.linspace(X.ppf(0.01),
                  X.ppf(0.99), 100)
#fda_binom = stats.binom.cdf(x, n, p) # Función de Distribución Acumulada
fda_normal = X.cdf(x_1) # Función de Distribución Acumulada
#plt.plot(x, fda_binom, '--', label='FDA binomial')
plt.plot(x_1, fda_normal, label='FDA nomal')
plt.title('Función de Distribución Acumulada')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.legend(loc=4)
plt.show()
```

```python
plt.plot(fda_normal, x_1,  label='FDA nomal')
plt.title('Función de Distribución Acumulada interpolar')
plt.xlabel('probabilidad')
plt.ylabel('valores')
plt.legend(loc=4)
plt.show()
```

```python
FDP_normal = X.pdf(x_1) # FDP
plt.plot(x_1, FDP_normal, label='FDP nomal')
plt.title('Función de Densidad de Probabilidad')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```
